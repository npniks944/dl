Problem Statement:

Assignment No. 1 Perform PCA in dimension reduction of numerical data

1. Pre-process the data through standardization.
2. Perform PCA to reduce dimension.
3. Construct the scree plot.
4. Data visualization in lower dimensional representation.

```python



import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA




data = pd.DataFrame({
    'Feature1': [10, 30, 50, 80, 110],
    'Feature2': [10, 15, 20, 25, 30],
    'Feature3': [5, 15, 25, 35, 45]
})

print("\nOriginal Dataset:")
print(data)




scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

scaled_df = pd.DataFrame(
    scaled_data,
    columns=data.columns
)

print("\nStandardized Dataset (Mean=0, Std=1):")
print(scaled_df)


# PCA

pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)

pca_df = pd.DataFrame(
    pca_data,
    columns=['Principal Component 1', 'Principal Component 2']
)

print("\nPCA Transformed Data (2 Components):")
print(pca_df)




print("\nExplained Variance Ratio:")
print(pca.explained_variance_ratio_)

print("\nCumulative Explained Variance:")
print(np.cumsum(pca.explained_variance_ratio_))


# Scree Plot

pca_full = PCA()
pca_full.fit(scaled_data)

plt.figure()
plt.plot(
    np.cumsum(pca_full.explained_variance_ratio_),
    marker='o'
)
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Scree Plot')
plt.grid(True)
plt.show()


#  2D PCA Visualization

plt.figure()
plt.scatter(
    pca_df['Principal Component 1'],
    pca_df['Principal Component 2']
)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA 2D Data Visualization')
plt.grid(True)
plt.show()
```
